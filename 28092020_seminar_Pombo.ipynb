### Complex multi-enhancer contacts captured by genome architecture mapping in *Nature*

#### Introduction
What we know about the structure of the nucleus is a result of microscopy and molecular methods for capturing spatial organization of the genome, e.g., chromatin conformation capture (3C)

But, these methods have limitations; genome architectsure mapping (GAM) is the first genome-wide method for capturing 3-D proximities between genomic loci without ligation and overcomes some limitations of 3C

#### Principle of the method
GAM combines "ultrathin cryosectioning with laser microdissection and DNA sequencing"

*What is cryosectioning?*
It might be when they freeze then cut a cell up by section

GAM infers parameters of chromatin spatial organization:
* genome-wide chromatin contact frequencies
* radial distributions
* chromatin compaction

Used mouse cells because there is a lot of information on chromatin occupancy already.

#### Identifying prominent interactions
SLICE, a general mathematical model, identifies the interactions most likely to be specific from GAM co-segregation data
* calculates a 'probability of interaction' (*Pi*) 
* applied genome-wide
* only analyzes 'prominently interaction' locus pairs (*i.e.*, *Pi* at a threshold of *P <= 0.05*

They tested whether active genes, enhancers, or inactive genes interact with each other more or less frequently than expected by chance; active genes and enhancers interact more, but inactive genes interact as expected by chance.

Overall, their results show that **interactions between regulatory sequences and active genes are major organizers of chromatin conformation**.

#### Detecting interacting triplets
*topologically associating domains* (**TADs**), large self-associating regions

GAM has a high enough resolution (at the level of hundreds of kilobases) to detect triplet contacts, which corresponds to the chromatin organization level of TADs

### Notes from seminar
*Cell-type specialization in the brain is encoded by specific longrange chromatin topologies*
Genome sequence code regulates gene expression at different scales
* At 100Mb, chromosome territories
* At 1-100Mb, hubs and compartments
* At 40kb-3Mb, TADs and loop domains
* At <1kb, promoter-enhancer contacts
    * RNA Pol II + enhancer
In medicine, an interest to understand genome code, since most DNA doesn't code for proteins, but some might decide if other DNA is being activated or not...
For example, 108 physically distinct loci implicated in schizophrenia. 
Disease-associated genetic variants are often found in non-coding regions: so which genes do they regulate?

Enhancers are highly cell-type specific: how do they work?
* Active promoters - more overlap among cerebellum, cortex, and subcortical structures
* Enhancers - significant variation among those cell-types

GAM: an unbiased approach to map chromatin contacts at all genomic scales
* Based on spatial statistics, similarly to other fields, ecology, sociology, etc.
* Each slice is a snapshot, so building up enough gives a pretty good picture of what's actually going on
    * Data, though, will be messy, so SLICE comes in to clean it up
* Procedure:
    * Sample is frozen, cryosection (200nm) by laser microdissection
    * Whole genome amplification
    * Next-generation sequencing
    * Calling DNA presence/absence
    * After about 200 samples, there is a pretty good picture of what's happening (reaches saturation)
* The technique is very flexible, so don't necessarily need to do a ton of samples for certain purposes -- collect as much data as needed for questions being asked!

Mapping 3D genome contacts in specialized cells in complex tissues with GAM
* In the brain, lots of different types of cells that are post-mitotic, they aren't reborn
* Immuno-GAM pipeline
    1. Whole tissue cryosectioning 
    2. Immunodetection of cells of interest
    3. Laser microdissection of cells of interest
    4. DNA extraction and sequencing 
    5. GAM data processing
* Technique is born from microscopy, which has been used for decades
* Looked at three different cell types
    1. Dopaminergic neurons
        * Reward-based learning
    2. Pyramidal glutamatergic neurons
        * Cognition, memory, frequently active
    3. Oligodendrocyte lineage cells
        * Myelination, neuron-glia communication
        * Recently post-mitotic
* Compare TADs, contact density and compaction, cell-type specific contacts, compartments A/B

GAM finds highly variable 3D topologies in brain cells, at short- and long-range genomic distances
* i.e., cells exhibit different internal structures in the nucleus
* means there are different interactions along chromatin
* only about 16Mb, diagrams on the slide, but chromosomes are >100Mb

Changes in 3D topology were integrated with cell-type specific gene expression obtained from scRNA-seq data
* extracted cells that they were isolating in the data set
* UMAP of single-cell transcriptomes - get cell types very different from one another
* used to determine which genes are on or off
* certain genes are expressed massively in one cell type over another

Extensive reorganization of topological domains (TADs) 
* Compared between different cell types
* Only 14% of TAD borders were conserved, most domains are highly dynamic
    * TAD borders vary by cell type and even within cell types
    * Are genes found in the borders?
    * By analyzing TAD borders, could find out what the cell type is.
    * So more active genes are found in TADs, and these active genes kind of define the cell

Very long neuronal genes undergo 'TAD melting' when most expressed
* TAD disappears when most expressed
* Nrxn3, a specific gene, is down-regulated when expressed, since it must be completely unwound when used (making it less compact)

* what did this loss of signal mean?
    * machine-learning approach, called prismer(?)
    * can have an ensemble with full range of solutions
    * can see how certain domains change in compactness from mESCs (mouse embryonic stem cells) to DNs (developed(?) neurons)
    * as the gene becomes highly expressed, get massive decompression of the gene being expressed
    * developed a strategy to calculate insulation score (IS) and then a map at multiple length scales (100-1000kb)
    * statistical test or similarity of IS value distributinos (genes > 400kb)
    * d and corresponding p-value are metric of dissimilarity of IS value distributions
    * TAD melting score = -log10(adj. p-value)
        * Syne1, Pde4d, Ecr2, Nrxn3, Ctnnd2 all exhibit high TAD melting scores

Genome-wide TAD melting at highly transcribed long genes
* melting appears in different conformations
* procedure captures these events, can appear in multiple places, with multiple TADs

*It is **not** the case that all genes that are expressed exhibit TAD melting*
* for example, Egr1 becomes more expressed as it is expressed, since it becomes more condensed (the domain)

#### Summary
* GAM can be applied in situ to specialized cells in complex tissues
* Genome architecture is highly cell-type specific in brain cells at multiple genomic scales
* Specfic 3D chromatin folding patterns reflect cell specialization and cell state
* Loss of domain (TAD) organization and 'TAD melting' at very long neuronal genes when most highly expressed
* Specific contacts at differentially expresse genes contain complex networks of TF motifs
